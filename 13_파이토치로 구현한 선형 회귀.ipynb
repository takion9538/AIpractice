```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
# 학습 데이터셋을 생성한다.
x_train = torch.FloatTensor([[1], [2], [3], [4]])
y_train = torch.FloatTensor([[50], [70], [90], [85]])
```

```python
# 단순 선형 회귀는 input=1, output=1
model = nn.Linear(1, 1) # 선형회귀 모델을 생성하는 클래스
```

```python
list(model.parameters())
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
epoch_count = 3000
for epoch in range(epoch_count + 1) : 
  H = model(x_train)

  cost = F.mse_loss(H, y_train) # 함수를 통해 평균 제곱 오차를 구한다.

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if epoch % 100 == 0 : 
    print(f'Epoch {epoch:4d}/{epoch_count}, Cost:{cost:.6f}')
```

```python
var = torch.FloatTensor([[5.0]])
pred = model(var)
print(f'입력이 5 일때의 예측값 : {pred.item()}')
```

```python
print(list(model.parameters()))
```

```python
x_train = torch.FloatTensor([[2, 0, 0], 
                             [2, 1, 0], 
                             [2, 2, 1], 
                             [3, 1, 1], 
                             [4, 1, 2], 
                             [4, 2, 2]])
y_train = torch.FloatTensor([[50], [60], [65], [70], [75], [85]]) 
```

```python
model = nn.Linear(3, 1) # 선형회귀 모델을 생성하는 클래스
```

```python
list(model.parameters())
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr=1e-2)
```

```python
epoch_count = 1000
for epoch in range(epoch_count + 1) : 

  H = model(x_train)
  cost = F.mse_loss(H, y_train)

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if epoch % 100 == 0 : 
    print(f'Epoch {epoch:4d}/{epoch_count}, Cost:{cost:.6f}')
```

```python
var = torch.FloatTensor([[5, 2, 3]])
pred = model(var)
print(f'예측값 : {pred}')
```
