### **1-1. Tensor**

* 텐서(Tensor) 는 배열이나 행렬과 매우 유사한 특수한 자료구조이다.
* Pytorch에서는 텐서를 사용하여 모델의 입력과 출력, 그리고 모델의 매개변수들을 부호화하는데 사용된다.

```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
print(x_data)
```

```python
import numpy as np
```

```python
np_array = np.array(data)
x_np_1 = torch.tensor(np_array) # ndarray 를 가져오는 것이 아니라 데이터를 복사해서 가져온다는 것을 확인.
print(x_np_1)
```

```python
x_np_1[0, 0] = 10
print(x_np_1) 
print(np_array)
```

```python
x_np_2 = torch.as_tensor(np_array) # 뷰( : 같은 데이터를 가리키는 포인터) 를 생성한다.
print(x_np_2)
```

```python
x_np_2[0, 0] = 10 
print(x_np_2) 
print(np_array)
```

```python
x_np_3 = torch.from_numpy(np_array) # 뷰 를 생성
print(x_np_3)
```

```python
x_np_3[0, 0] = 20
print(x_np_3) 
print(np_array)
```

```python
# 텐서를 ndarray 로 변경
np_again = x_np_1.numpy()
print(type(np_again))
```

```python
a = torch.ones(2, 3)
print(a)

b = torch.zeros(2, 3)
print(b)

c = torch.full((2, 3), 2)
print(c)

d = torch.empty(2, 3)
print(d)
```

```python
i = torch.eye(3) # 카탈로그화
print(i)
```

```python
j = torch.arange(10) # 0 ~ x 까지의 숫자를 나열
print(j)
```

```python
k = torch.rand(2, 2)
print(k)
l = torch.randn(2, 2) # 가우시안 정규분포 : 평균을 0으로 하고, 표준편차가 1인 랜덤값 (-2 ~ 2)
print(l)
```

### **1-2. Tensor 속성**

```python
tensor = torch.rand(3, 4)

print(f'Shape of tensor : {tensor.shape}') # 텐서의 형태
print(f'DataType of tensor : {tensor.dtype}') # 텐서의 데이터타입
print(f'Device of tensor : {tensor.device}') # 텐서를 계산할 때 사용되는 디바이스 : gpu / cpu
```

```python
# 속성 변경
tensor = tensor.reshape(4, 3)
tensor = tensor.int()
if torch.cuda.is_available() : 
  tensor = tensor.to('cuda')

print(f'Shape of tensor : {tensor.shape}')
print(f'DataType of tensor : {tensor.dtype}')
print(f'Device of tensor : {tensor.device}')
```

### **1-3. Indexing 과 Slicing**

```python
a = torch.arange(1, 13).reshape(3, 4)
print(a)
```

```python
print(a[1])
print(a[0, -1])
```

```python
print(a[1 : -1])
```

```python
print(a[:2, 2:])
```

### **1-4. Transpose**

```python
a = torch.arange(16).reshape(2, 2, 4)
print(a, a.shape)
```

```python
b = a.transpose(1, 2)
print(b, b.shape)
```

```python
c = a.permute((2, 0, 1)) # 텐서플로우의 transpose 와 쓰임이 똑같다.
print(c, c.shape)
```

### **1-5. Tensor 연산**

```python
x = torch.tensor([[1, 2], [3, 4]], dtype=torch.float32)
y = torch.tensor([[5, 6], [7, 8]], dtype=torch.float32)
print(x)
print(y)
```

```python
print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x @ y)
print('*' * 30)
print(torch.add(x, y))
print(torch.subtract(x, y))
print(torch.multiply(x, y))
print(torch.divide(x, y))
print(torch.matmul(x, y)) # 연산 결과는 같다.
```

```python
# in - place 연산
print(x.add(y))
print(x) # 기존의 값이 유지된다.

print(x.add_(y)) # x 에 결과를 다시 저장
# add : x = x + 10, add_ : x += 10
print(x)
```

```python
z = torch.arange(1, 11).reshape(2, 5)
print(z)
```

```python
sum1 = torch.sum(z, axis=0)
sum2 = torch.sum(z, axis=1)
sum3 = torch.sum(z, axis=-1)
print(sum1, sum1.shape)
print(sum2, sum2.shape)
print(sum3, sum3.shape)
```

```python
a = torch.arange(24).reshape(4, 6)
b = a.clone().detach() # clone : 복사하기, detach : 붙여넣기. 둘 다 사용해야 함
print(a, a.shape)
print(b, b.shape)
```
