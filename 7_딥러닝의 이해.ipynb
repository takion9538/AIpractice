# **1. 딥러닝의 진화**

### **2012**
* CNN 중 Alex net 라는 네트워크를 구축하였다.
* Alex net 이 1등을 차지하였다. -> CNN 이 유명해지게 된다.

### **2013**
* 아타리 게임에 AI 를 적용하여 학습을 통해 구석을 파서 공을 띄우는 방법을 깨달았다.
* 딥마인드가 구축되고, 구글에 인수되어 알파고가 탄생했다.

### **2014**
* RNN 이라는 네트워크를 사용하여 중국어를 영어로 번역하였다.
* 성능의 한계가 생겼다.
* attention 모델의 출현으로 성능이 급격히 좋아졌다.

### **2015**
* GANs 에서 오바마의 합성 사진을 배포한다.
* Residual Nexworks(ResNet) : CNN 의 종류중 하나이다.
* 사람과 Ai(ResNet) 이 이미지를 찾는 테스트를 진행하였다. 
* 사람은 평균적으로 5% 의 오차율이 발생하였다.
* ResNet 은 3% 의 오차율이 발생하였다.

### **2016**
* 알파고 : 이세돌 9단이 컴퓨터를 이긴 마지막 인류가 되었다.

### **2017**
* RNN 의 단점을 극복하였다.
* attention 만으로 만든 모델 Transformer(번역모델) 이 탄생하였다.
* 자연어를 정복하였다.
* 딥러닝에서 가장 중요한 모델 중의 하나이다.

### **2018**
* Transformer 에서 인코더만을 따온 모델 BERT 가 구축되었다.
* 자연어를 정답 없이 사용한다. (인터넷 문장들의 데이터를 삽입하였다.)
* 일부 글자를 가리고 가린 부분을 맞출 수 있도록 학습하였다.
* 모델을 크게 만들고, 엄청난 데이터를 사용하였다.

### **2019/2020**
* GPU 수천대를 이용하여 큰 모델을 만든다.
* 정답이 필요하지 않아 실제 데이터만으로 셀프 수퍼바이즈 러닝이 가능하다.

# **2. 퍼셉트론(Perceptron)**

* Perception : 인지하는 능력
* Neuron : 감각 입력 정보를 의미있는 정보로 바꿔주는 뇌에 있는 신경세포
* Perceptron : Perception + Neuron 로, 생물학적 뉴런이 감각 정보를 받아 문제를 해결하는 원리를 본따 만든 인공 뉴런이다.

# **3. 딥러닝의 키 포인트**

* 데이터
* 모델(CNN, RNN, Transformer, 멀티레이어 퍼셉트론 등)
* 알고리즘(Gradient Descent 를 기초로 많은 알고리즘이 만들어졌다.)
* loss function : 데이터가 얼마나 기울기(정답률)와 떨어져 있는지 (오차율)

# **4. Neural Network 에서의 핵심**

* Deep Neural Network : layer 수가 최소 2개 이상인 네트워크
* Loss Function / Cost Function : Neural Network 가 얼마나 성능이 좋거나 나쁜지에 대한 척도
* loss 값을 줄이는 방법 : 미분(접선의 기울기) 를 이용
> 적절한 weight 를 찾는 것이 핵심이다.
